```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
IMG_ETRI_532 = plt.imread('IMG_ETRI_532_2048x2048.png')
IMG_ETRI_780 = plt.imread('IMG_ETRI_780_2048x2048.png')

IMG_loss0_532 = plt.imread('IMG_loss0_532_2048x2048.png')
IMG_loss0_780 = plt.imread('IMG_loss0_780_2048x2048.png')

IMG_loss1_532 = plt.imread('IMG_loss1_532_2048x2048.png')
IMG_loss1_780 = plt.imread('IMG_loss1_780_2048x2048.png')
```

```python
IMG_stack_532 = np.hstack((IMG_ETRI_532, IMG_loss0_532, IMG_loss1_532))
IMG_stack_780 = np.hstack((IMG_ETRI_532, IMG_loss0_532, IMG_loss1_532))
```

```python
plt.figure(figsize=(21,7))
plt.imshow(IMG_stack_532, cmap='gray')
plt.title('532nm')
plt.axis('off')
plt.show()

plt.figure(figsize=(21,7))
plt.imshow(IMG_stack_780, cmap='gray')
plt.title('780nm')
plt.axis('off')
plt.show()
```
